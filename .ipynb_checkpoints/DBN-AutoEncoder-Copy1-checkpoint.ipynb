```python
from keras.layers import Input, Dense, Convolution2D, MaxPooling2D, UpSampling2D
from keras.models import Model
from keras import regularizers
import numpy as np
import scipy.io as sio
WB=sio.loadmat('WB.mat')['WB']
x_train = WB
x_test = WB
x_train = x_train.astype('float32') / 1.
x_test = x_test.astype('float32') / 1.


# this is our input placeholder
input_img = Input(shape=(1, 200, 200))

x = Convolution2D(24, 6, 6, activation='relu', border_mode='same')(input_img)
encoded = MaxPooling2D((2, 2), border_mode='same')(x)

encoder_1st = Model(input=input_img, output=encoded)

# x = UpSampling2D((2, 2))(encoded)
# decoded = Convolution2D(1, 6, 6, activation='relu', border_mode='same')(x)

decode_pool=UpSampling2D((2, 2))
decode_conv=Convolution2D(1, 6, 6, activation='relu', border_mode='same')

x=decode_pool(encoded)
decoded=decode_conv(x)

decoder_input = Input(shape=(24,100,100))
_h_decoded = decode_pool(decoder_input)
_x_decoded = decode_conv(_h_decoded)
generator_1st = Model(decoder_input, _x_decoded)

autoencoder_1st = Model(input_img, decoded)
autoencoder_1st.compile(optimizer='adadelta', loss='mse')

# decoder_layer = autoencoder.layers[-1]
# # create the decoder model
# decoder = Model(input=input_img, output=decoder_layer(input_img))


print x_train.shape
print x_test.shape
x_train=x_train.reshape(100,1,200,200)
x_test=x_test.reshape(100,1,200,200)
print x_train.shape
print x_test.shape
```

```python
autoencoder_1st.summary()
```

```python
autoencoder_1st.fit(x_train, x_train,
                nb_epoch=200,
                batch_size=10,
                shuffle=True,
                validation_data=(x_test, x_test))
```

```python
autoencoder_1st.save_weights('model_weights_layer1.h5')
autoencoder_1st.load_weights('model_weights_layer1.h5',by_name=True)
```

```python
encoded_imgs = encoder_1st.predict(x_test)
generat_imgs = generator_1st.predict(encoded_imgs)
recon_1st = autoencoder_1st.predict(x_test)
```
